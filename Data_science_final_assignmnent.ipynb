##having benefits

```python

def having(benfits):

  totalpay,totalpaybenefits=benfits
  if totalpay==totalpaybenefits:
    return 'have no benefits'
  else:
    return 'have  benefits'
```

```python
salary['having_benefits']=salary[['TotalPayBenefits','TotalPay']].apply(having,axis=1)
salary
```

```python
sns.countplot(x='having_benefits',data=salary,hue='Year')
```

```python
sns.countplot(x='having_benefits',data=salary)
```

#working over time

```python
def have(overtime):

 
  if overtime==0:
    return 'have no overtime'
  else:
    return 'have  overtime'
```

```python
salary['overtime']=salary['OvertimePay'].apply(have)
salary
```

```python
sns.countplot(x='overtime',data=salary)
```

```python
sns.countplot(x='overtime',data=salary,hue='Year')
```

```python
sns.countplot(x='overtime',data=salary,hue='having_benefits')
```
